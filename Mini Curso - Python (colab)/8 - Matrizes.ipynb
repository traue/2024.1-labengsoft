Resolução 1:

```
lista = []
for _ in range(1, 6):
  valor = int(input('Digite o valor: '))
  lista.append(valor)
```

```
lista
```

```
len(lista)
```

```
soma = 0
for i in range(len(lista)):
  #print(lista[i])
  soma += lista[i]
print('Soma: ', soma)
```

```
#outra forma de fazer
import numpy as np
np.array(lista).sum()
```

```
matriz = np.array([[3, 4, 1],
                   [3, 1, 5]])
matriz
```

```
matriz.shape
```

```
soma = 0
for i in range(matriz.shape[0]):
  for j in range(matriz.shape[1]):
    #print(matriz[i][j])
    soma += matriz[i][j]
print('Soma: ', soma)
```
