## Operadores lógicos

```
a = True
b = False
```

```
print(a, b)
```

```
a and b
```

```
a & b #mesmo que o bloco acima
```

```
c = a and b
print("'A' e 'B' são iguais é", c) # para enfatizar o 'A'e o 'B' Usar "" quando tiver que usar ' dentro 
```

```
a or b
```

```
a | b #mesmo que o acima
```

```
d = a or b
print("'A' ou 'B' é igual a", d)
```

```
not a
```

```
not b
```
